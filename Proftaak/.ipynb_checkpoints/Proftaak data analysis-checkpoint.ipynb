## CL_BOUNDS_CT_MDS to PL_QUO_CT_TRANSACTNOW
These tables describe how much (bound) policies were quoted through another *unit*.
Since these go per 2, we can seperate them from the rest and look at which unit is doing the "best". There is missing data within these columns, all these 9999 values can be replaced with 0 without making much of an inpact.
## Loss ratio and Loss Ratio 3YR
These values seem to be the loss percentages that the company in questions have gotten. This data could be worked into a label by taking all positive numbers (losses) as 0 and negative numbers (gains) as 1. There are also 0s in there, this could be that the company was on the breakeven point or that this value is missing. To use this column as a label we will have to delete all the missing data in it. Filtering these values would leave us at 55946 rows out of 190571, a big loss of data.
## Growth rate 3YR
**There needs to be 4 FULL years of data for this number to be computed**. The increase in the amount of premiums written within those years. Could be usefull for indication and the amount of missing data is manageable. 108941 out of 190751 records have data within them. A very usable column.

## Retention Ratio
Too many missing values, excel bugs out and doesn't recognise it but I think at least 50-75% is missing. Meaning of the table is awkward and weird. Not useful to us
## PRD_INCRD_LOSSES_AMT
Losses. 14625 out of 190571 records left when filtered. No information on what this table means. Unusable.


# Conclusion
**Growth rate** will be a very important column to look at seeing its size and meaningful data within it.

**Loss ratio** is a nice backup to have but has a lot of missing values.

**PRD_INCRD_LOSSES_AMT** has a lot of values that could be used but are currently meaningless.

**Retention Ratio** weird data with lots of missing values without much of a meaning.

**Garbage data at the end** usefull when looking at outside sources but not that meaningful on its own
